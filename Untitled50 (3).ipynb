```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset
from transformers import get_linear_schedule_with_warmup, AdamW
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold
import optuna
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score, confusion_matrix, ConfusionMatrixDisplay

# Load the dataset
try:
    df = pd.read_csv("/content/air_data_set.csv")
except FileNotFoundError:
    raise Exception("The dataset file was not found. Please check the file path.")

# Convert timestamp to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'], format='%d-%m-%Y %H:%M')

# Extract additional time-based features
df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.dayofweek

# Create lagged features
for lag in range(1, 25):
    df[f'PM2.5_lag_{lag}'] = df['PM2.5'].shift(lag)

# Drop rows with NaN values (from lagged features)
df = df.dropna().reset_index(drop=True)

# Separate features and target variable
X = df.drop(columns=['timestamp', 'PM2.5'])
y = df['PM2.5']

# Split data into training and test sets
test_size = 0.2
split_index = int(len(X) * (1 - test_size))
X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define sequence length
sequence_length = 24

class TimeSeriesDataset(Dataset):
    def __init__(self, data, target, sequence_length):
        self.data = data
        self.target = target
        self.sequence_length = sequence_length

    def __len__(self):
        return len(self.data) - self.sequence_length

    def __getitem__(self, index):
        x = self.data[index:index+self.sequence_length]
        y = self.target[index+self.sequence_length]
        return torch.tensor(x, dtype=torch.float32), torch.tensor(y, dtype=torch.float32)

class TransformerTimeSeries(nn.Module):
    def __init__(self, input_dim, embed_dim, num_heads, num_layers, dropout=0.1):
        super(TransformerTimeSeries, self).__init__()
        self.embedding = nn.Linear(input_dim, embed_dim)
        encoder_layer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=num_heads, dropout=dropout)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.fc = nn.Linear(embed_dim, 1)

    def forward(self, x):
        x = self.embedding(x)
        x = x.permute(1, 0, 2)  # Transpose for transformer [seq_len, batch_size, embed_dim]
        x = self.transformer_encoder(x)
        x = x.mean(dim=0)  # Average pooling
        x = self.fc(x)
        return x

def train_and_evaluate(model, train_loader, val_loader, optimizer, scheduler, num_epochs=10):
    train_losses = []
    val_losses = []

    for epoch in range(num_epochs):
        model.train()
        epoch_train_loss = 0
        for inputs, targets in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = nn.MSELoss()(outputs.squeeze(), targets)
            loss.backward()
            optimizer.step()
            scheduler.step()
            epoch_train_loss += loss.item()
        train_losses.append(epoch_train_loss / len(train_loader))

        model.eval()
        epoch_val_loss = 0
        with torch.no_grad():
            for inputs, targets in val_loader:
                outputs = model(inputs)
                loss = nn.MSELoss()(outputs.squeeze(), targets)
                epoch_val_loss += loss.item()
        val_losses.append(epoch_val_loss / len(val_loader))

    return train_losses, val_losses

def predict(model, X_test, sequence_length):
    model.eval()
    test_dataset = TimeSeriesDataset(X_test, np.zeros(len(X_test)), sequence_length)
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
    predictions = []
    with torch.no_grad():
        for inputs, _ in test_loader:
            outputs = model(inputs)
            predictions.append(outputs.squeeze().cpu().numpy())
    predictions = np.concatenate(predictions, axis=0)
    return predictions

def calculate_rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

# Hyperparameter optimization using Optuna
def objective(trial):
    num_heads = trial.suggest_int('num_heads', 2, 8)
    embed_dim = trial.suggest_int('embed_dim', 16, 128)

    num_layers = trial.suggest_int('num_layers', 1, 4)
    dropout = trial.suggest_float('dropout', 0.1, 0.5)
    lr = trial.suggest_loguniform('lr', 1e-4, 1e-2)

    try:
        model = TransformerTimeSeries(input_dim=X_train_scaled.shape[1], embed_dim=embed_dim, num_heads=num_heads, num_layers=num_layers, dropout=dropout)
    except AssertionError:
        return float('inf')

    optimizer = AdamW(model.parameters(), lr=lr)

    kf = KFold(n_splits=5)
    results = []

    for train_index, val_index in kf.split(X_train_scaled):
        X_train_split, X_val_split = X_train_scaled[train_index], X_train_scaled[val_index]
        y_train_split, y_val_split = y_train.iloc[train_index].reset_index(drop=True), y_train.iloc[val_index].reset_index(drop=True)

        train_dataset = TimeSeriesDataset(X_train_split, y_train_split, sequence_length)
        val_dataset = TimeSeriesDataset(X_val_split, y_val_split, sequence_length)

        train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
        val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

        total_steps = len(train_loader) * 10
        scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=total_steps)

        train_loss, val_loss = train_and_evaluate(model, train_loader, val_loader, optimizer, scheduler)
        results.append(val_loss[-1])  # Use the final validation loss

    average_loss = np.mean(results)
    return average_loss

# Optimize hyperparameters
sequence_length = 24
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=4)
best_params = study.best_params

print("Best parameters: ", best_params)

# Train the final model with the best parameters
model = TransformerTimeSeries(input_dim=X_train_scaled.shape[1], embed_dim=best_params['embed_dim'], num_heads=best_params['num_heads'], num_layers=best_params['num_layers'], dropout=best_params['dropout'])
optimizer = AdamW(model.parameters(), lr=best_params['lr'])
total_steps = len(DataLoader(TimeSeriesDataset(X_train_scaled, y_train, sequence_length), batch_size=32, shuffle=True)) * 10
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=total_steps)

kf = KFold(n_splits=5)
train_losses, val_losses = [], []

for train_index, val_index in kf.split(X_train_scaled):
    X_train_split, X_val_split = X_train_scaled[train_index], X_train_scaled[val_index]
    y_train_split, y_val_split = y_train.iloc[train_index].reset_index(drop=True), y_train.iloc[val_index].reset_index(drop=True)

    train_dataset = TimeSeriesDataset(X_train_split, y_train_split, sequence_length)
    val_dataset = TimeSeriesDataset(X_val_split, y_val_split, sequence_length)

    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

    train_loss, val_loss = train_and_evaluate(model, train_loader, val_loader, optimizer, scheduler)
    train_losses.extend(train_loss)
    val_losses.extend(val_loss)

# Plot training and validation loss
plt.figure(figsize=(12, 6))
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()

# Make predictions
predictions = predict(model, X_test_scaled, sequence_length)

# Performance Metrics
y_true = y_test.values
y_pred = predictions

# Calculate MSE, RMSE, and R-squared
#mse = mean_squared_error(y_true, y_pred)
#rmse = calculate_rmse(y_true, y_pred)
#r2 = r2_score(y_true, y_pred)

#print(f"Mean Squared Error (MSE): {mse}")
#print(f"Root Mean Squared Error (RMSE): {rmse}")
#print(f"R-squared (R2): {r2}")

# Plot actual vs predicted values
plt.figure(figsize=(12, 6))
plt.plot(np.arange(len(y_true)), y_true, label='True Values', color='blue')
plt.plot(np.arange(len(y_pred)), y_pred, label='Predicted Values', color='red')
plt.xlabel('Index')
plt.ylabel('PM2.5')
plt.title('True vs Predicted PM2.5 Values')
plt.legend()
plt.show()
```
